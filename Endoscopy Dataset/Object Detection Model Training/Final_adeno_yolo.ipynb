```python
model = YOLO('/content/object_detection/best_adeno_M_416.pt')
```

```python
img = Image.open('/content/object_detection/adeno_images/adeno_a (1).jpg')
resized_image = img.resize((350, 350))
results = model(resized_image)
class_name = results[0].names
clss = class_name[0]
boxes = results[0].boxes
box = boxes[0]
confidence = box.conf
conf = str(confidence.item())
bbox = box.xyxy
bbox_list = bbox.tolist()[0]
```

```python
print(clss)
print(conf)
print(bbox_list)
```
